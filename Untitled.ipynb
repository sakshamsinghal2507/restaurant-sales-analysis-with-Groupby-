```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style("whitegrid")
df=sns.load_dataset("tips")
df.head()
```

```python
df.groupby("day", observed=True)["total_bill"].sum()
```

```python
df.groupby("sex", observed=True)["total_bill"].mean()
```

```python
df.groupby("time", observed=True)["tip"].sum()
```

```python
df.groupby("day", observed=True)["total_bill"].sum().plot(kind="bar")
plt.title("Total sales per day")
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Bill vs Tip")
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation heatmap")
plt.show()
```

```python
# Insights:

# 1. Some days have higher total sales.
# 2. Spending pattern differs between male and female customers.
# 3. Dinner generates more tips than lunch.
# 4. Higher bill leads to higher tip.
# 5. Strong relation between total_bill and tip.
```
